```python
from sklearn.model_selection import ParameterSampler
from scipy.stats import uniform
parametros= {'eps': uniform(0,7) ,'min_samples':[5,10,15], 'metric': ['euclidean', 'manhattan', 'mahalanobis']}
```

```python
lista=list(ParameterSampler(parametros,n_iter= 15))
lista
```

```python
def evaluacion_estimador(data_frame, lista):
    resultados=[]
    for parametros in lista:
        estimador=DBSCAN(**parametros)                       
        labels=estimador.fit_predict(data_frame)
        score=silhouette_score(data_frame,labels)
        resultados.append([f'{parametros} con puntuacion', score])
    return resultados
```

```python
resultados=evaluacion_estimador(df_normalizado, lista=lista)
```

```python
sorted(resultados, key= lambda x: x[1], reverse=True)[:5]
```

```python
mejores_parametros={'eps': 5.132890839566071, 'metric': 'euclidean', 'min_samples': 10}
estimador=DBSCAN(**mejores_parametros)
```

```python
etiquetas=estimador.fit_predict(df_normalizado)
```

```python
pd.Series(etiquetas).unique()
```

```python
pd.Series(etiquetas).value_counts()
```

```python
def resumen_cluster(cluster_id):
    cluster = df[etiquetas==cluster_id]
    resumen_cluster=cluster.mean().to_dict()
    resumen_cluster["cluster_id"] = cluster_id
    return resumen_cluster

def comparar_clusters(*cluster_ids):
    resumenes = []
    for cluster_id in cluster_ids:
        resumenes.append(resumen_cluster(cluster_id))
    return pd.DataFrame(resumenes).set_index("cluster_id").T
```

```python
comparar_clusters(0,-1)
```
